```python
# Mengimpor modul drive dari pustaka google.colab, yang memungkinkan akses ke Google Drive.
from google.colab import drive

# Mengimpor modul pandas dan menggantinya dengan alias 'pd' untuk manipulasi dan analisis data.
import pandas as pd

# Mengimpor modul pyplot dari pustaka Matplotlib dan menggantinya dengan alias 'plt' untuk membuat visualisasi.
import matplotlib.pyplot as plt

# Mengimpor modul seaborn dan menggantinya dengan alias 'sns' untuk membuat grafik statistik yang lebih menarik.
import seaborn as sns

# Mengimpor modul widgets dari pustaka ipywidgets untuk membuat widget interaktif.
import ipywidgets as widgets

# Mengimpor fungsi 'interact' dari modul ipywidgets untuk membuat widget interaktif.
from ipywidgets import interact

# Mengimpor modul numpy dan menggantinya dengan alias 'np' untuk operasi numerik.
import numpy as np
```

```python
path = 'drive/MyDrive/DatasetML/diamonds.csv'

df = pd.read_csv(path)

print(df.head())
```

```python
# Menggunakan metode info() pada dataframe df
df.info()
```

```python
# Menampilkan statistik fitur
print("Statistik untuk setiap fitur:")
print(df.describe(include='all'))
```

```python
# Mendefinisikan daftar fitur yang ingin diperiksa untuk jumlah nilai unik
features = ['carat', 'cut', 'color', 'clarity', 'depth', 'table', 'price', 'x', 'y', 'z']

# Iterasi melalui setiap fitur dalam daftar fitur
for feature in features:
    # Menggunakan metode nunique() untuk menghitung jumlah nilai unik dalam kolom fitur
    # dan mencetak hasilnya dengan format yang sesuai
    print(f"Jumlah unique value for {feature}: {df[feature].nunique()}")
```

```python
# Mengelompokkan numerical features
numerical_features = ['carat', 'depth', 'table', 'x', 'y', 'z']

# Mengelompokkan categorical features
categorical_features = ['cut', 'color', 'clarity']

# Mengelompokkan target
target = ['price']

# Menampilkan hasil pengelompokkan
print("Numerical Features:", numerical_features)
print("Categorical Features:", categorical_features)
print("Target:", target)
```

```python
# Membuat bar chart untuk setiap fitur kategorikal
for feature in categorical_features:
    # Membuat figur dengan ukuran tertentu
    plt.figure(figsize=(8, 6))

    # Menggunakan sns.countplot untuk membuat diagram batang
    sns.countplot(x=feature, data=df, palette='viridis')

    # Menambahkan judul diagram batang
    plt.title(f'Bar Chart of {feature}', size=15)

    # Menambahkan label sumbu x dan y
    plt.xlabel(feature, size=12)
    plt.ylabel('Count', size=12)

    # Menampilkan diagram batang
    plt.show()
```

```python
# Membuat pie chart untuk setiap fitur kategorikal
for feature in categorical_features:
    # Membuat figur dengan ukuran tertentu
    plt.figure(figsize=(8, 6))

    # Menggunakan value_counts() untuk menghitung jumlah kemunculan setiap kategori
    # Lalu membuat pie chart dengan plot(kind='pie')
    # Menggunakan autopct untuk menampilkan persentase dalam bentuk angka desimal
    # Mengatur warna potongan pie chart dengan colors
    df[feature].value_counts().plot(kind='pie', autopct='%1.1f%%', colors=['skyblue', 'lightgreen', 'lightcoral', 'orange', 'lightblue'])

    # Menambahkan judul pie chart
    plt.title(f'Pie Chart of {feature}', size=15)

    # Menghilangkan label sumbu y agar tidak terlihat
    plt.ylabel('')

    # Menampilkan pie chart
    plt.show()
```

```python

# Mengelompokkan numerical features
numerical_features = ['carat', 'depth', 'table', 'price', 'x', 'y', 'z']

# Menghitung matriks korelasi antar fitur numerik
correlation_matrix = df[numerical_features].corr()

# Membuat heatmap dari matriks korelasi
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Heatmap Korelasi antar Fitur Numerik')
plt.show()
```

```python
# Mengelompokkan numerical features
numerical_features = ['carat', 'depth', 'table', 'x', 'y', 'z']

# Membuat distribusi untuk setiap fitur numerik
for feature in numerical_features:
    # Membuat figur dengan ukuran tertentu
    plt.figure(figsize=(8, 6))

    # Menggunakan seaborn (sns) untuk membuat histogram (histplot) dari fitur numerik
    # dengan tambahan KDE (Kernel Density Estimation) untuk menggambarkan estimasi kepadatan
    sns.histplot(df[feature], kde=True, color='darkcyan')

    # Menambahkan judul pada plot
    plt.title(f'Distribution of {feature}', size=15)

    # Menambahkan label sumbu x
    plt.xlabel(feature, size=12)

    # Menambahkan label sumbu y
    plt.ylabel('Frequency', size=12)

    # Menampilkan plot distribusi
    plt.show()
```

```python
# Fungsi untuk menampilkan distribusi dengan interaktif bin width
def plot_histogram(feature, bin_width):
    # Membuat gambar (figure) dengan ukuran 8x6
    plt.figure(figsize=(8, 6))
    # Menghitung batasan bin dengan menggunakan np.arange berdasarkan nilai minimum dan maksimum fitur ditambah dengan lebar bin
    bins = np.arange(min(df[feature]), max(df[feature]) + bin_width, bin_width)
    # Membuat histogram dengan Seaborn, menentukan batasan bin (bins), menambahkan kurva KDE, dan mengatur warna histogram
    sns.histplot(df[feature], bins=bins, kde=True, color='darkcyan')
    # Menambahkan judul ke plot
    plt.title(f'Distribution of {feature}', size=15)
    # Menambahkan label sumbu x dengan nama fitur
    plt.xlabel(feature, size=12)
    # Menambahkan label sumbu y dengan teks "Frequency"
    plt.ylabel('Frequency', size=12)
    # Menampilkan plot
    plt.show()

# Menampilkan slider interaktif untuk setiap fitur numerik
for feature in numerical_features:
    # Menggunakan fungsi interact dari library widgets untuk membuat slider interaktif
    interact(plot_histogram, feature=widgets.fixed(feature), bin_width=(0.1, 1, 0.1))
```

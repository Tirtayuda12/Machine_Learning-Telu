```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# Menggunakan Pandas untuk membaca data dari file CSV
df = pd.read_csv('/content/drive/MyDrive/DatasetML/HRDataset_v14.csv')

# Tampilkan beberapa baris pertama dataset
print(df.head())
```

```python
# Mengecek info dari dataframe

df.info()
```

```python
df.head()
```

```python
# Mengubah tipe data Gender ID dengan tipe data string untuk mengubah 1 menjadi laki - laki dan 0 menjadi perepuan
# Buat mapping untuk penggantian nilai
mapping = {1: 'laki-laki', 0: 'perempuan'}

# Gunakan fungsi replace dengan mapping
df['GenderID'] = df['GenderID'].replace(mapping)

# Mengubah nama label GenderID dengan Gender
df = df.rename(columns={'GenderID': 'Gender'})

```

```python
df.head()
```

### Scatter plot

```python
# Scatter plot antara variabel independen (misalnya Age) dan variabel dependen (Salary)
plt.figure(figsize=(8, 6))
plt.scatter(df['PerfScoreID'], df['Salary'], alpha=0.5)
plt.xlabel('PerfScoreID')
plt.ylabel('Salary')
plt.title('PerfScoreID vs. Salary')
plt.show()
```

### Diagram Batang

```python
# Diagram batang untuk melihat banyaknya karyawan tiap departemen
plt.figure(figsize=(10, 6))
sns.countplot(x='Department', data=df, palette='viridis')
plt.title('Banyaknya karyawan tiap Departement')
plt.xlabel('Department')
plt.ylabel('karyawan')
plt.xticks(rotation=45)
plt.show()
```

### Histogram

```python
# Boxplot gaji berdasarkan departemen untuk melihat range gaji terhadap departemen
plt.figure(figsize=(12, 8))
sns.boxplot(x='PerfScoreID', y='Salary', data=df)
plt.title('Boxplot Gaji Berdasarkan Skor Performa Karyawan')
plt.xlabel('Skor Performa')
plt.ylabel('Gaji')
plt.show()
```

```python
# Boxplot gaji berdasarkan departemen
plt.figure(figsize=(12, 8))
sns.boxplot(x='Department', y='Salary', data=df)
plt.title('Boxplot Gaji Berdasarkan Departemen')
plt.xlabel('Departemen')
plt.ylabel('Gaji')
plt.show()
```

### Pie Chart

```python
### Diagram pie gender
plt.figure(figsize=(8, 8))
df['Gender'].value_counts().plot(kind='pie', autopct='%1.1f%%', colors=['lightblue', 'lightcoral'])
plt.title('Sebaran dari Gender')
plt.ylabel('')
plt.show()
```

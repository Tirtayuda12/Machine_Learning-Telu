## Exploratory Data Analysis (EDA)

Exploratory Data Analysis (EDA) adalah proses analisis awal yang dilakukan pada dataset untuk memahami karakteristik data, mengidentifikasi pola, menemukan anomali, dan merumuskan hipotesis yang dapat diuji lebih lanjut. Berikut adalah langkah-langkah umum dalam melakukan EDA:

1. Memahami Dataset:
  * Baca Dokumentasi: Pelajari informasi tentang dataset, termasuk penjelasan setiap kolom (fitur) dan nilai-nilai yang mungkin.
  * Identifikasi Jenis Data: Tentukan tipe data untuk setiap kolom (numerik, kategori, string, dll.).

2. Pemeriksaan Data Awal:
  * Tampilkan Data: Lihat beberapa baris pertama dataset untuk memahami struktur dan format data.
  * Periksa Informasi: Gunakan df.info() (dengan df adalah DataFrame Pandas) untuk melihat informasi tentang jumlah entri, tipe data, dan apakah ada nilai yang hilang.
3. Membersihkan Data:
  * Periksa dan Tangani Nilai Kosong: Identifikasi dan tangani nilai kosong atau null. Anda bisa menghapus baris atau mengisi nilai yang hilang dengan metode seperti interpolasi atau imputasi.
  * Penanganan Anomali: Identifikasi dan tangani nilai yang tidak mungkin atau anomali dalam data.
4. Analisis Statistik Deskriptif:
  * Ringkasan Statistik: Gunakan df.describe() untuk melihat statistik deskriptif seperti mean, median, kuartil, dll. untuk kolom-kolom numerik.
  * Distribusi Data: Visualisasikan distribusi data menggunakan histogram, boxplot, atau KDE plot.
5. Explorasi Variabel:
  * Univariat Analysis: Analisis variabel secara individual untuk memahami distribusi dan pola data.
  * Bivariat Analysis: Analisis hubungan antara dua variabel. Gunakan scatter plot untuk melihat korelasi antar variabel numerik dan boxplot untuk membandingkan distribusi berdasarkan kategori.
6. Visualisasi Data:
  * Histogram: Untuk melihat distribusi variabel numerik.
  * Boxplot: Untuk melihat distribusi, kuartil, dan outlier.
  * Barplot: Untuk melihat frekuensi atau perbandingan antara kategori.
  * Scatter plot: Untuk melihat hubungan antar variabel numerik.
7. Pengelompokan dan Agregasi:
  * Pivot Tables: Untuk mengelompokkan data dan melihat statistik agregat berdasarkan beberapa fitur.
  * Groupby: Untuk melakukan pengelompokan data berdasarkan kategori tertentu dan menghitung statistik di setiap grup.
8. Analisis Korelasi:
  * Heatmap: Untuk visualisasi matriks korelasi antar variabel numerik.
  * Pairplot: Untuk melihat scatter plot dan distribusi variabel secara bersamaan.
9. Penyajian Informasi:
  * Gunakan narasi dan visualisasi yang jelas untuk menyajikan temuan dan insight dari analisis data.
  * Buat kesimpulan dan hipotesis yang dapat diuji lebih lanjut.
10. Iterasi dan Pemrosesan Ulang:
  * Setelah melakukan EDA awal, mungkin Anda perlu kembali ke langkah-langkah sebelumnya, melakukan pembersihan tambahan, atau mengeksplorasi aspek data yang lebih dalam.
  
  EDA adalah langkah awal yang penting sebelum membangun model atau melakukan analisis lanjutan. Ini membantu memahami data secara menyeluruh dan membimbing pengambilan keputusan yang tepat dalam analisis data.

Setelah dilihat menggunakan method info, ada sebanyak 1253 baris dan 22 kolom pada dataset IT salary EU report 2020.

Tiap kolom juga terdapat banyak kekeliruan tipe data contohnya tipe data pada kolom total years of experience yang harusnya numerik namun setelah di cek bertipe data objek

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Statistik deskriptif
print(df.describe())

# Memeriksa missing values
print(df.isnull().sum())

# Visualisasi distribusi gaji
plt.figure(figsize=(10, 6))
sns.histplot(df['Salary'], bins=30, kde=True, color='skyblue')
plt.title('Distribution of IT Salaries in EU 2020')
plt.xlabel('Salary (EUR)')
plt.ylabel('Frequency')
plt.show()

# Perbandingan gaji berdasarkan gender
plt.figure(figsize=(10, 6))
sns.boxplot(x='Gender', y='Salary', data=df)
plt.title('IT Salary Distribution by Gender in EU 2020')
plt.xlabel('Gender')
plt.ylabel('Salary (EUR)')
plt.show()

# Perbandingan gaji berdasarkan negara
plt.figure(figsize=(12, 8))
sns.barplot(x='Country', y='Salary', data=df, estimator=np.median, ci=None)
plt.title('Median IT Salary by Country in EU 2020')
plt.xlabel('Country')
plt.ylabel('Median Salary (EUR)')
plt.xticks(rotation=45, ha='right')
plt.show()

# Korelasi antara gaji, pengalaman, dan pendidikan
plt.figure(figsize=(8, 6))
sns.heatmap(df[['Salary', 'YearsExperience', 'EducationLevel']].corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix')
plt.show()
```
